```python
#МАШИНЫ

import pandas as pd
import random
import math
import time

start_time = time.time()

buyers = pd.DataFrame()

data = pd.read_csv("Car_sales.csv")

data['параметр 0'] = data['Price_in_thousands']
data['параметр 1'] = data['Horsepower']
data['параметр 2'] = data['Fuel_efficiency']

buyers.insert(0, "параметр 0", round(data['параметр 0']))
buyers.insert(1, "параметр 1", round(data['параметр 1'] * 0.1))
buyers.insert(2, "параметр 2", round(data['параметр 2'] * 0.1))

buyers = buyers.dropna()
buyers = buyers.reset_index(drop=True)

qual_amount = buyers.shape[1] - 1
```

```python
#Подсчёт количества покупателей при данном векторе товара
def amount_of_buyers(fr_name, vector):
    n = fr_name.shape[0]
    counter = 0
    for j in range (0, n):
        condition = (vector[0] <= fr_name['параметр 0'][j])
        for i in range(1, qual_amount + 1):
            name = 'параметр ' + str(i)
            condition &= vector[i] >= fr_name[name][j]
        if condition:
            counter += 1
    return counter

#Подсчёт прибыли
def get_profit(fr_name, vector):
    profit = vector[0]
    for i in range(1, qual_amount + 1):
        profit -= vector[i]
    profit = profit * amount_of_buyers(fr_name, vector)
    return profit
```

```python
# Реализация алгоритма заметающей прямой для одного качества
if (qual_amount == 1):
    prices = list(sorted(set(buyers['параметр 0'].values.tolist()), reverse = True))
    parametr1 = list(sorted(set(buyers['параметр 1'].values.tolist()), reverse = True)) 
    d = {}
    list2_1 = parametr1.copy()
    def price_par1(fr_name, list1, list2):
        candidate = [0, 0]
        for m in range (0, len(list1)):
            candidate[0] = list1[m]
            k = 0
            for k in range (0, len(list2)):
                list2_2 = list2.copy()
                candidate[1] = list2[k]
                a = get_profit(fr_name, candidate)
                key = candidate[1]
                if d.get(key, 1) == 1:                
                    d[key] = a
                elif a >= d[key]:
                    d[key] = a
                elif a < d[key]:
                    list2_2.remove(candidate[1])
            list2 = list2_2            
        return(d)

    price_par1(buyers, prices, parametr1)
    print(d)
    max_val = max(d.values())
    print("--- %s seconds ---" % (time.time() - start_time))
    print(max_val)
    
# Реализация алгоритма заметающей прямой для двух качеств
if (qual_amount == 2):
    prices = list(sorted(set(buyers['параметр 0'].values.tolist()), reverse = True))
    parametr1 = list(sorted(set(buyers['параметр 1'].values.tolist()), reverse = True))
    parametr2 = list(sorted(set(buyers['параметр 2'].values.tolist()), reverse = True))
    d = {}
    list2_1 = parametr1.copy()
    def price_par1(fr_name, list1, list2, list3):
        candidate = [0, 0, 0]
        for l in range (0, len(list3)):
            candidate[2] = list3[l]
            list2 = list2_1
            for m in range (0, len(list1)):
                candidate[0] = list1[m]
                k = 0
                for k in range (0, len(list2)):
                    list2_2 = list2.copy()
                    candidate[1] = list2[k]
                    a = get_profit(fr_name, candidate)
                    key = candidate[1]
                    if d.get(key, 1) == 1:                
                        d[key] = a
                    elif a >= d[key]:
                        d[key] = a
                    elif a < d[key]:
                        list2_2.remove(candidate[1])
                list2 = list2_2            
        return(d)

    price_par1(buyers, prices, parametr1, parametr2)
    max_val = max(d.values())
    print(max_val)
    print("--- %s seconds ---" % (time.time() - start_time))
```
