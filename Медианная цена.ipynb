```python
import time

start_time = time.time()

# Ввод данных
import pandas as pd

n = 0
m = 0

#Подсчет количества элементов в файле "Продавцы"
for line in open('sellers.txt', 'r'):
    n += 1

#Подсчет количества элементов в файле "Покупатели"
for line in open('buyers.txt', 'r'):
    m += 1

#Объявление списков продавцов и покупателей
sell_list = [] 
buy_list = []

#Заполнение списка продавцов
with open('sellers.txt','r') as file:
    for line in file:
        price = int((line))
        sell_list.append(price)
        
if m > n:
    for i in range(n, m):
        sell_list.append(float('inf'))
        
#Заполнение списка покупателей
with open('buyers.txt','r') as file:
    for line in file:
        price = int((line))
        buy_list.append(price)
        
if n > m:
    for k in range(m, n):
        buy_list.append(int('0'))

#Объединение списков
list1 = sell_list + buy_list

#нахождение медианы за среднее время O(n)
list_of_pivots = []
import random
def quickselect_median(l, pivot_rand=random.choice):
    if len(l) % 2 == 1:
        return quickselect(l, len(l) / 2, pivot_rand)
    else:
        return (quickselect(l, len(l) / 2 - 1, pivot_rand), quickselect(l, len(l) / 2, pivot_rand))

def quickselect(l, k, pivot_rand):
    
    pivot = pivot_rand(l)
    list_of_pivots.append(pivot)

    lows = [i for i in l if i < pivot]
    highs = [i for i in l if i > pivot]
    pivots = [i for i in l if i == pivot]

    if k < len(lows):
        return quickselect(lows, k, pivot_rand)
    elif k < len(lows) + len(pivots):
        return pivots[0]
    else:
        return quickselect(highs, k - len(lows) - len(pivots), pivot_rand)
    
(a, b) = quickselect_median(list1, pivot_rand=random.choice)


#Подсчет пар, вступивших в сделку
counter1 = 0
counter2 = 0

buy_success = []
sell_success = []

for i in range(max(m,n)):
    if buy_list[i] >= a:
        counter1 += 1
        buy_success.append(buy_list[i])
    if sell_list[i] <= a:
        counter2 += 1
        sell_success.append(sell_list[i])
        
#Количество пар, вступающих в сделку должно быть одинаково, иначе не будет равновесия
if counter1>counter2:
    buy_success.remove(buy_success[counter1-1])
if counter2>counter1:
    sell_success.remove(sell_success[counter2-1])
        
amount_of_pairs = min(counter1,counter2)

print('Медианная цена:')
if a == b:
    print(a,'\n')
else:
    print(a, b,'\n')

print('Количество пар, вступающих в сделку:')
print(amount_of_pairs, '\n')

# print('Пары, вступающие в сделку:')
# print(buy_success)
# print(sell_success)
    
print("--- %s seconds ---" % (time.time() - start_time))
```
